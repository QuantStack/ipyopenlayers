```python
with open("demo.json") as f:
    data = json.load(f)

g = GeoJSON(data=data)
m.add_layer(g)
```

```python
g.style = {
    'strokeColor': '#1966ad',       # Bleu Foncé pour le contour
    'strokeWidth': 2,               # Largeur du contour
    'pointRadius': 7,               # Rayon moyen pour les points
    'fillColor': '#e3eef8',         # Bleu Très Clair pour le remplissage
    'pointFillColor': '#8cc0dd',    # Bleu Moyen pour le remplissage des points
    'pointStrokeColor': '#084b93',  # Bleu Nuit pour le contour des points
    'pointStrokeWidth': 1.5         # Largeur du contour des points
}
```
